## Read dataset and remove headers

```python
rdd1 = sc.textFile('npr.csv').cache()
rdd2 = sc.textFile('abc.csv').cache()
rdd3 = sc.textFile('cbs.csv').cache()
rdd4 = sc.textFile('cnn.csv').cache()
rdd5 = sc.textFile('foxfriends.csv').cache()
rdd6 = sc.textFile('foxnews.csv').cache()
rdd7 = sc.textFile('la.csv').cache()
rdd8 = sc.textFile('nbc.csv').cache()
rdd9 = sc.textFile('bbc.csv').cache()

# Function to remove the header.
def remove_header(x):
    header = x.first()
    return x.filter(lambda row: row != header)

rdd1 = remove_header(rdd1)
rdd2 = remove_header(rdd2)
rdd3 = remove_header(rdd3)
rdd4 = remove_header(rdd4)
rdd5 = remove_header(rdd5)
rdd6 = remove_header(rdd6)
rdd7 = remove_header(rdd7)
rdd8 = remove_header(rdd8)
rdd9 = remove_header(rdd9)
```

## Merge into one RDD

```python
data = sc.union([rdd1, rdd2, rdd3, rdd4, rdd5, rdd6, rdd7, rdd8, rdd9])
data.take(2)
```

## Split into columns
Only retrieve name, message, description, and captions.

```python
data2 = data.map(lambda x : x.lower().split(',')[2:6])
```

```python
data2.take(2)
```

## Filter posts with the word "Thai" in it

```python
posts = "thai"
```

```python
data3 = data2.filter(lambda x: posts in x[0])
```

```python
data3.collect()
```

### Number of posts which are related to Thailand

```python
num_post = data3.collect()
print(len(num_post))
```

## Now that we have all the posts that are related to Thailand, we can find popular nouns among them

```python
data3 = data3.flatMap(lambda x: x)
data3 = data3.flatMap(lambda x: x.split(" "))
```

```python
data3.take(2)
```

## Keep only nouns

```python
words = data3.collect()
str1 = ' '.join(str(e) for e in words)
```

```python
# Clean up the data some more.
import re
import string
def clean_text(text):
    # Remove numbers
    text_nonum = re.sub(r'\d+', '', text)
    # Remove punctuations and convert characters to lower case
    text_nopunct = "".join([char.lower() for char in text_nonum if char not in string.punctuation]) 
    # Substitute multiple whitespace with single whitespace
    # Also, removes leading and trailing whitespaces
    text_no_doublespace = re.sub('\s+', ' ', text_nopunct).strip()
    return text_no_doublespace
str1 = clean_text(str1)
```

### Using nltk, tokenize each word and then use pos_tags to select only nouns

```python
import nltk
nouns = [word for (word, pos) in nltk.pos_tag(nltk.word_tokenize(str1)) if pos[0] == 'N']
print(nouns[0:10])
```

### Convert back to RDD

```python
data3 = sc.parallelize(list(nouns))
data3.take(2)
```

## Count popular nouns

```python
data3 = data3.map(lambda x : (x,1))
from operator import add
data3 = data3.reduceByKey(add)
data3 = data3.map(lambda x: (x[1], x[0]))
data3 = data3.sortByKey(ascending=False)
data3.collect()
```
